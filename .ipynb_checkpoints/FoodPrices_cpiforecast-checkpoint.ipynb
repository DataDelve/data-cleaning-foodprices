```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import xlsx data file
df = pd.read_excel('CPIForecast.xlsx')
```

```python
df.info()
df.head()
```

```python
cols = ['item', 'relative_importance', 'mon-to-mon_aug2023_sept2023', 'yr-over-yr_sept2022_sept2023', 
        'ytd_avg_2023_to_avg_2022', 'annual_2021', 'annual_2022', '20-yr_hist_avg_2003_to_2022',
        'forecast_2023_lower', 'forecast_2023_mid', 'forecast_2023_upper', 'forecast_2024_lower',
        'forecast_2024_mid', 'forecast_2024_upper']
```

```python
# Fix column names, using data from row 0
df.columns = cols
df.info()
df.head()
```

```python
# drops the top row with duplicate column names
df = df.drop([0])

# drops any row where 'Consumer Price Index item' is NaN, because that row is invalid.
df = df[df['item'].notna()]
```

```python
# resets index, after we dropped several rows
df = df.reset_index(drop=True)
```

```python
df
```

```python
# drops rows 22 through the end of the table, because the data is not relevant
df = df.drop(df.index[22:])
```

```python
df
```

```python
df.to_csv('CPIForecast_cleaned.csv', index=False)
```
